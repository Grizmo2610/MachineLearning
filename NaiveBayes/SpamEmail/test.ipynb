```python
import numpy as np
import pandas as pd
```

```python
path = './ex6DataPrepared/'
train_data_fn = 'train-features.txt'
train_label_fn = 'train-labels.txt'
test_data_fn = 'test-features.txt'
test_label_fn = 'test-labels.txt'

def readData(data_fn, label_fn):
    with open(path + label_fn, 'r') as f:
        label = f.readlines()
    label = list(map(int, [x.strip() for x in label]))

    with open(path + data_fn, 'r') as file:
        content = [list(map(int, line.strip().split())) for line in file]

    data = [[0 for i in range(2500)] for i in range(len(label))]

    for part in content:
        data[part[0] - 1][part[1] - 1] += part[2]

    return data, label


train_data, train_label = readData(train_data_fn, train_label_fn)

print(len(train_data))
print(len(train_label))

result = [[] for i in range(700)]
for i in range(700):
    result[i] = [train_label[i]]
    result[i].extend(train_data[i])

test = np.array(result)

df = pd.DataFrame(test)

df.rename(columns={df.columns[0]: 'Is spam'}, inplace=True)

for i in range(1, 2501):
    df.rename(columns={df.columns[i]: 'W' + str(i)}, inplace=True)
```
